```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/ks-projects-201801.csv")
data_df = pd.DataFrame(data)
data_df
```

**Подсчет категорий по популярности**

```python
category_counts = data_df['category'].value_counts()
top_10_categories = category_counts.head(10)
others_count = category_counts.iloc[10:].sum()  # Суммируем остальные категории
top_10_categories['Другие'] = others_count  # Добавляем категорию "Другие"

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))  # Настройка размера графика
top_10_categories.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.tab20.colors, 
    textprops={'fontsize': 10}
)
plt.title('Распределение проектов по категориям (топ-10)', fontsize=14)
plt.ylabel('')  # Убираем метку оси Y для эстетики
plt.tight_layout()

# Сохранение графика (если нужно)
# plt.savefig('top_10_categories_pie_chart.png')

plt.show()
```

**Успех по категориям**

```python
successful_counts = data_df[data_df['state'] == 'successful']['category'].value_counts()

total_counts = data_df['category'].value_counts()

success_rate = (successful_counts / total_counts * 100).sort_values(ascending=False)

if 'Другие' in success_rate.index:
    success_rate = success_rate.drop('Другие')

top_10_success = success_rate.head(10)

plt.figure(figsize=(12, 6))
top_10_success.plot(kind='bar', color='green', edgecolor='black')
plt.title('Топ-10 категорий с наивысшим процентом успешных проектов', fontsize=16)
plt.xlabel('Категории', fontsize=12)
plt.ylabel('Процент успешных проектов', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Сохранение графика (если нужно)
# plt.savefig('top_10_successful_categories.png')

plt.show()
```
